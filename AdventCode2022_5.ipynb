Assumptions
 - Stacks are always named 1 to n in order.
 - Move orders never try to take more crates than are on a stack.
 - Move orders are always from and to valid stacks

## Part 1 solution with tracing output

```python
from urllib.request import urlopen
input_source = "https://raw.githubusercontent.com/MathewGarvin/Advent2022/main/advent2022day5input.txt"
f = urlopen(input_source)

x = 1
t = []

while x:
    a = f.readline().decode()
    t.append(a[1::4])
    x = a[1] != '1'
    
t = [''.join(s).strip() for s in zip(*t)]
a = f.readline()
a = f.readline()

print(t)

# for i in range(5):
while a != b'':
    print(a.split()[1::2])
    b = [int(d) for d in a.split()[1::2]]
    print("moving", t[b[1] - 1][b[0] - 1::-1], "to stack", b[2])
    t[b[2] - 1] = t[b[1] - 1][b[0] - 1::-1] + t[b[2] - 1]
    t[b[1] - 1] = t[b[1] - 1][b[0]:]

    print(t)
    a = f.readline()
    
f.close()
print(''.join([d[0] for d in t]))
```

## Parts 1 & 2 combined solution

```python
from urllib.request import urlopen
s = "https://raw.githubusercontent.com/MathewGarvin/Advent2022/main/advent2022day5input.txt"

def F(p):
    f = urlopen(s)
    a = s
    t = []

    while a[1] != '1':
        a = f.readline().decode()
        t.append(a[1::4])

    t = [''.join(s).strip() for s in zip(*t)]
    
    for d in '12':
        a = f.readline()

    while a != b'':
        b = [int(d) for d in a.split()[1::2]]
        c = b[1] - 1
        e = b[2] - 1
        t[e] = (1 - p) * t[c][b[0]-1::-1] + p * t[c][0:b[0]] + t[e]
        t[c] = t[c][b[0]:]
        a = f.readline()

    print(''.join([d[0] for d in t]))

F(0)
F(1)
```
